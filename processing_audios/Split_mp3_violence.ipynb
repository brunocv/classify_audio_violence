```python
from pydub import AudioSegment
```

# Split mp3 violence

```python
def split(file, time):
    sound = AudioSegment.from_mp3("../dataset/violence/"+ str(file) + ".mp3")

    # pydub does things in milliseconds
    seconds = time * 1000

    first_half = sound[:seconds]
    second_half = sound[seconds:]

    first_half.export("../dataset/nonviolence/" + str(file) + "_2.mp3", format="mp3")
    second_half.export("../dataset/violence/"+ str(file) + ".mp3", format="mp3")
```
