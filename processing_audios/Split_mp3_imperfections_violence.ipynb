# Split mp3 violence

```python
def split_violence(file, time):
    sound = AudioSegment.from_mp3("../dataset/violence/"+ str(file) + ".mp3")

    # pydub does things in milliseconds
    seconds = time * 1000

    first_half = sound[:seconds]
    second_half = sound[seconds:]

    first_half.export("../dataset/violence/" + str(file) + ".mp3", format="mp3")
    second_half.export("../dataset/remover_violence/"+ str(file) + ".mp3", format="mp3")
```
